```python
import selenium.webdriver as webdriver
from selenium.webdriver.chrome.service import Service
import requests
from bs4 import BeautifulSoup

def scrape_website(website):
    try:
        result = requests.get(website).text
    except Exception as e:
        result = f"Something went wrong: {str(e)}"
    return result

def extract_body_content(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    body_content = soup.body
    if body_content:
        return str(body_content)
    return "HTML Body was not found."

def clean_body_content(body_content):
    soup = BeautifulSoup(body_content, "html.parser")

    for script_style in soup(["script", "style"]):
        script_style.extract()

    cleaned_content = soup.get_text(separator="\n")
    cleaned_content = "\n".join(
        line.strip() for line in cleaned_content.splitlines() if line.strip()
    )

    return cleaned_content

#LLM models usually have a limit of words that can process at once. That limit is around 8000 tokens
#In this function we will be splitting our cleaned text into batches that we will proceed to fit into the model one by one
def split_dom_content(dom_content, max_length=6000):
    return [
        dom_content[i : i + max_length] for i in range(0, len(dom_content), max_length)
    ]


```

```python
from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate

template = (
    "You are tasked with extracting specific information from the following text content: {content}. "
    "Please follow these instructions carefully: \n\n"
    "1. **Extract Information:** Only extract the information that directly matches the provided description: {criteria}. "
    "2. **No Extra Content:** Do not include any additional text, comments, or explanations in your response. "
    "3. **Empty Response:** If no information matches the description, return an empty string ('')."
    "4. **Direct Data Only:** Your output should contain only the data that is explicitly requested, with no other text."
)



model = OllamaLLM(model="llama3.1")

def parse(batches, criteria):
    prompt = ChatPromptTemplate.from_template(template)
    #go to prompt first then pass it to the model
    chain = prompt | model

    parsed_results = []

    for i, batch in enumerate(batches, start=1):
        response = chain.invoke({"content":batch, "criteria":criteria})
        print(f"Parsed Batch : {i} of {len(batches)}")
        parsed_results.append(response)

    return response

```

```python
URL = "https://en.wikipedia.org/wiki/1975_Pacific_hurricane_season#Season_summary"
criteria = "for each hurricane you are required to collect its name, its start and end date, number of deaths it caused and areas it affected. These criteria should be searched for each hurricane separately and the results for each hurricane should be returned in dictionary form that follows the template {hurricane_name : hurricane_name_value, start_date:hurricane_start_date, end_date:hurricane_end_date, human_casualties:hurricane_deahs_value, areas_affected:List containing all areas affected by the hurricane }"
result = scrape_website(URL)
body_content = extract_body_content(result)
cleaned_body_content = clean_body_content(body_content)
print(cleaned_body_content)
batches = split_dom_content(cleaned_body_content)
parsed_results = parse(batches, criteria)
print(f"results : {parsed_results}")
```
